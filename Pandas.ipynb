### Select a Pandas Series from a Data Frame

```python
UFO_Reports.head()
```

```python
# To Select a series we have 2 ways. 1- dot(.)- notation-Not works in all cases; 2- brackets[] notation

type(UFO_Reports['City'].head()) # Accessing the City column from data frame. Type is series
```

```python
#[] notation works in any scenario
UFO_Reports['City'].head()
```

```python
#dot(.) notation. If any column name is Python builtin function name then . notation considers the column name as function name
# Also if the column names are separated with space we can't access that column using this notation

UFO_Reports.City.head()
```

```python
UFO_Reports['Colors Reported'].head()
```

```python
#Adding a new series to the Data Frame
# Create a new column:Location to the Data Frame

UFO_Reports['Location'] = UFO_Reports.City + ', ' + UFO_Reports.State
```

### Applying Multiple Filter Criteria to a Pandas DataFrame

```python
Movies_Ratings.head()
```

```python
Movies_Ratings[Movies_Ratings.Duration >= 200]
```

```python
Movies_Ratings[(Movies_Ratings['Duration'] >= 200) & (Movies_Ratings.Genre == 'Crime')]
```

```python
Movies_Ratings[(Movies_Ratings['Duration'] >= 220) | (Movies_Ratings.Genre == 'Drama')].head()
```

```python
# Other simple way to define multiple OR conditions

Movies_Ratings[Movies_Ratings.Genre.isin(['Crime', 'Drama', 'Action'])].head()
```
